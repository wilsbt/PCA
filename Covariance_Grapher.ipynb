Notebook written to show various ways of presenting a covariance matrix.

```python
import numpy as np

A = [45, 37, 42, 35, 39]
B = [38, 31, 26, 28, 33]
C = [10, 15, 17, 21, 12]

data = np.array([A, B, C])

covMatrix = np.cov(data, bias=False)
print(covMatrix)
```

```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt

A = [45, 37, 42, 35, 39]
B = [38, 31, 26, 28, 33]
C = [10, 15, 17, 21, 12]

data = np.array([A, B, C])

cov_matrix = np.cov(data,bias=True)
sn.heatmap(cov_matrix, annot=True, fmt='g')
plt.show()
```

If we use a large dataset

```python
n = 100 # number of variables
rng = np.random.RandomState(1) # seed 
X = np.dot(rng.rand(n, n), rng.randn(n, 200))
# print(X)
print(np.shape(X)) # (2, 200) means we have 2 rows and 200 columns
```

```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt

cov_matrix = np.cov(X)
sn.heatmap(cov_matrix, annot=False, fmt='g')
plt.show()
```
