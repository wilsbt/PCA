This notebook is for finding eigenvalues and eigenvectors of matrices, for specific application of covariance matrices.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Make Some data

```python
rng = np.random.RandomState(585) # seed 
X = np.dot(rng.rand(2, 2), rng.randn(2, 200))
print(np.shape(X)) # (2, 200) means we have 2 rows and 200 columns
```

```python
plt.scatter(X[0], X[1])
plt.show()
```

Find covariance matrix

```python
C = np.cov(X, bias=True)
print(C)
```

To find the eigenvalues and eigenvectors of this matrix, numpy functions can be called.

```python
evals, evecs = np.linalg.eig(C)
evecs = evecs.T # this line is to make the rows of evecs the eigenvectors
print("The eigenvalues are:", evals)
print("The eigenvectors are:", evecs[0], evecs[1])
```

```python
plt.scatter(X[0], X[1])
plt.arrow(x=0, y=0, dx=evecs[0,0], dy=evecs[1,1], head_width=0.2, color='k')
plt.arrow(x=0, y=0, dx=evecs[1,0], dy=evecs[1,1], head_width=0.2, color='k')
plt.show()
```

# Standardized data

```python
Xm = np.mean(X, axis=1) # mean
Xs = np.std(X, axis=1) # standard deviation
Xv = np.var(X, axis=1) # variance
Xn = ((X.T-Xm.T)/Xs.T).T # standardization
print(np.mean(Xn, axis=1))
print(np.std(Xn, axis=1)) # expect [1, 1]
print(np.var(Xn, axis=1)) # expect [1, 1]
```

```python
plt.scatter(Xn[0], Xn[1])
plt.show()
```

```python
C = np.cov(Xn, bias=False)
print(C)
```

```python
evals, evecs = np.linalg.eig(C)
evecs = evecs.T # this line is to make the rows of evecs the eigenvectors
print("The eigenvalues are:", evals)
print("The eigenvectors are:", evecs[0], evecs[1])
```

```python
plt.scatter(X[0], X[1])
plt.arrow(x=0, y=0, dx=evecs[0,0], dy=evecs[1,1], head_width=0.2, color='k')
plt.arrow(x=0, y=0, dx=evecs[1,0], dy=evecs[1,1], head_width=0.2, color='k')
plt.show()
```
